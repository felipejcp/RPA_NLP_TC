```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
from nltk.util import ngrams
from nltk.corpus import stopwords
from collections import Counter
import itertools
import pandas as pd

df = pd.read_excel('!arquivo_para_analise.xlsx',sheet_name='filtro 4 - pt para en')

documentos = df['Summary List']



# Função para extrair bigrams de um texto e remover stopwords
def extract_bigrams_and_remove_stopwords(text):
    text = text.replace("-", "")
    tokens = [word for word in text.split() if word.lower() not in stopwords.words()]
    bigrams = list(ngrams(tokens, 1))
    return [" ".join(bigram) for bigram in bigrams]

# Extrair bigrams e remover stopwords de todos os documentos
all_bigrams = list(itertools.chain.from_iterable([extract_bigrams_and_remove_stopwords(doc) for doc in documentos]))


# Contar a frequência dos bigrams
bigram_counter = Counter(all_bigrams)

# Extrair os 30 primeiros bigrams mais frequentes
top_bigrams = bigram_counter.most_common(30)

# Separar bigrams e contagens
bigrams, contagens = zip(*top_bigrams)

plt.figure(figsize=(10, 8))
plt.barh(bigrams, contagens, color='gray')  # Define a cor para preto
plt.xlabel("Frequency",fontsize=16) 
plt.ylabel("Bigrams",fontsize=16)
plt.gca().invert_yaxis()
plt.gca().set_facecolor('white')  # Define o fundo como branco
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.barh(bigrams, contagens, color='gray')  # Define a cor para preto
plt.xlabel("Frequency",fontsize=16) 
plt.ylabel("Unigrams",fontsize=16)
plt.gca().invert_yaxis()
plt.gca().set_facecolor('white')  # Define o fundo como branco
plt.show()
```
